```python
import pandas as pd
import os
```

```python
anos = [2000, 2002, 2004, 2006, 2008, 2010, 2012, 2015, 2017, 2019, 2022, 2024]

colunas_desejadas = ['Location', 'Time', 'AgeGrp', 'PopMale', 'PopFemale', 'PopTotal']
```

```python
df_frames = []

for ano in anos:
    arquivo = f'WPP{ano}_PopulationByAgeSex_Medium.csv'
    caminho_arquivo = os.path.join('Pop_bases', arquivo)
    df = pd.read_csv(caminho_arquivo, usecols=colunas_desejadas, low_memory=False)
    df['Ano'] = ano
    df_frames.append(df)

    
df_final = pd.concat(df_frames, ignore_index=True)

df = []
```
